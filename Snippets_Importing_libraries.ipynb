```
from manim import *
```

```
%%manim -qm -v WARNING SquareToCircle

class SquareToCircle(Scene):
   def construct(self):
      square = Square()
      circle = Circle()
      circle.set_fill(PINK, opacity=0.5)
      self.play(Create(square))
      self.play(Transform(square, circle))
      self.wait()
```

```
from manim import *

class Counting(Scene):
    def construct(self):
        apples = VGroup(*[
            Circle(radius=0.3).shift(RIGHT*i)
            for i in range(5)
        ])

        self.play(LaggedStart(*[Create(a) for a in apples]))
        self.play(Write(Text("1 2 3 4 5").scale(1.2).to_edge(DOWN)))
        self.wait()
```

```
class MultiplicationArray(Scene):
    def construct(self):
        dots = VGroup()

        for row in range(4):
            for col in range(3):
                dots.add(Dot().shift(RIGHT*col + DOWN*row))

        self.play(Create(dots))
        self.play(Write(MathTex("4 \\times 3 = 12")))
```

```
from manim import *

class MathShort(Scene):
    def construct(self):
        title = Text("Learn Fractions", font_size=60)
        self.play(Write(title))
        self.wait()

        frac = MathTex(r"\frac{1}{2} + \frac{1}{2} = 1")
        self.play(Transform(title, frac))
        self.wait()

        self.play(frac.animate.scale(1.5))
        self.wait()
```
